```julia
using Plots

function rand_X_y()
    return rand([1,-1]) * rand(), rand([1,-1]) * rand()
end

function in_circle(x, y)
    return sqrt(x ^ 2 + y ^ 2) < 1
end


N = 10000
C = 0

circle_x = []
circle_y = []
square_x = []
square_y = []

for i = 1:N
    x, y = rand_X_y()
    if in_circle(x, y)
        global C += 1
        append!(circle_x, x)
        append!(circle_y, y)
    else
        append!(square_x, x)
        append!(square_y, y)
    end
end

PI = (4 * C) / N

println(PI)

# println(circle)
# println(square)

gr()
# plot(circle, square)
plot(square_x, square_y, seriestype = :scatter, color="red")
plot!(circle_x, circle_y, seriestype = :scatter, color="blue")
```
